```python
import matplotlib.pyplot as plt

def plot_iterations():
    # Datos
    node_sets = {
        "50 Nodos": ([0.04, 0.06, 0.1], [8.852, 6.928, 5.408]),
        "100 Nodos": ([0.02, 0.04, 0.05], [10.904, 6.96, 6.298]),
        "200 Nodos": ([0.01, 0.02, 0.03], [13.236, 8.264, 6.34])
    }

    # Crear figuras
    plt.figure(figsize=(12, 4))

    for i, (title, (densities, iterations)) in enumerate(node_sets.items(), 1):
        plt.subplot(1, 3, i)
        plt.plot(densities, iterations, marker='o', linestyle='-', color='b')
        plt.xlabel("Densidad")
        plt.ylabel("Número medio de iteraciones")
        plt.title(title)
        plt.grid(True)

    plt.tight_layout()
    plt.show()

# Ejecutar la función
plot_iterations()
```
